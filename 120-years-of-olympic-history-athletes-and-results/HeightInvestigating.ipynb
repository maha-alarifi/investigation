### What are the agverege hieght, highest and shorrtest for each gender since 1920? 

```python
maskYear = df["Year"]>= 1920
maskMale = df["Sex"]=='M'
maskFemale = df["Sex"]=='F'
maskAge = df["Age"]>=24
```

```python
df2 = df[maskYear]
df2.shape
```

```python
df3 = df2[maskAge]
df3.shape
```

```python
plt.plot(df3[maskMale].groupby('Year').max().Height);
plt.plot(df3[maskMale].groupby('Year').mean().Height);
plt.plot(df3[maskMale].groupby('Year').min().Height);
plt.title("Avrage hieghts among years|Men")
plt.axhline(240)
plt.show();
```

```python
plt.plot(df3[maskFemale].groupby('Year').max().Height);
plt.plot(df3[maskFemale].groupby('Year').mean().Height);
plt.plot(df3[maskFemale].groupby('Year').min().Height);
plt.title("Avrage hieghts among years|Women")
plt.axhline(240)
plt.show();
```

### Which countries tend to have highst players?

```python
df3[maskFemale].groupby('NOC').mean().Height.sort_values(ascending=False).index[0:10]
```

```python
from matplotlib.pyplot import figure
figure(num=None, figsize=(5, 4), dpi=80, facecolor='w', edgecolor='k')
plt.bar(df3[maskFemale].groupby('NOC').mean().Height.sort_values(ascending=False).index[0:10], df3[maskFemale].groupby('NOC').max().Height.sort_values(ascending=False)[0:10])
plt.xticks(rotation=90);
plt.title("Top 10 coutries with highest women heights average")
```

```python
figure(num=None, figsize=(5, 4), dpi=80, facecolor='w', edgecolor='k')
plt.bar(df3[maskMale].groupby('NOC').max().Height.sort_values(ascending=False).index[0:10], df3[maskMale].groupby('NOC').max().Height.sort_values(ascending=False)[0:10])
plt.xticks(rotation=90);
plt.title("Top 10 coutries with highest men heights average")
```
